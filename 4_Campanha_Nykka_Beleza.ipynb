KPIs para a mineração de dados:

*   **Receita:** É o valor total de vendas gerado, ou seja, todo o dinheiro que a empresa arrecada com a venda de seus produtos e serviços. É a base para avaliar o faturamento.

*   **Lucro:** É o dinheiro que sobra após subtrair todos os custos e despesas da receita. Representa o verdadeiro ganho financeiro da empresa e sua capacidade de gerar riqueza.

*   **Valor Gasto:** Refere-se ao investimento total feito em marketing e campanhas para promover os produtos ou serviços. Mostra o quanto estamos investindo para alcançar nossos objetivos.

*   **ROI (Retorno Sobre Investimento):** Média o retorno financeiro em relação ao investimento. Ele mostra o quanto a empresa lucra para cada real investido, indicando a eficiência geral do investimento.

*   **ROAS (Retorno Sobre o Gasto com Anúncio):** Mede a receita gerada por cada real gasto especificamente em anúncios. É um indicador direto da eficácia das campanhas de publicidade em gerar vendas.

*   **Taxa de Conversão:** É a porcentagem de visitantes ou leads que realizaram uma ação desejada, como uma compra. Indica a eficiência em transformar o interesse em um resultado concreto.

*   **CPA (Custo por Aquisição de Cliente):** É o custo médio para adquirir um novo cliente ou uma nova venda. Essencial para entender o custo-benefício de cada aquisição de cliente.

*   **Ticket Médio:** É o valor médio que um cliente gasta em cada compra. Um ticket médio alto significa que os clientes estão comprando mais por transação, aumentando a receita por venda.

*   **CPL (Custo por Lead):** É o custo médio para gerar um lead, ou seja, um contato de um potencial cliente que demonstrou interesse. Indica o custo de atrair pessoas interessadas em nossos produtos.

*   **CPC (Custo por Clique):** É o valor médio pago por cada clique em um anúncio. É uma métrica importante para avaliar a eficiência do investimento em publicidade digital e o custo da atenção do público.

*   **CTR (Taxa de Cliques):** É a porcentagem de pessoas que clicaram em um anúncio em relação ao número total de vezes que ele foi exibido. Um CTR alto indica que o anúncio é relevante e atraente para o público.

*   **Impressoes:** É o número total de vezes que um anúncio foi exibido aos usuários, independentemente de ter sido clicado ou não. Representa a visibilidade e o alcance da campanha.

*   **Cliques:** É o número de vezes que os usuários interagiram com um anúncio clicando nele. Indica o interesse inicial do público no conteúdo do anúncio.

*   **Leads:** São potenciais clientes que demonstraram interesse em um produto ou serviço, por exemplo, preenchendo um formulário ou navegando por uma página de vendas. São contatos qualificados para futuras vendas.

*   **Conversoes:** É o número de vendas ou outras ações desejadas que foram realizadas pelos usuários. É o objetivo final da maioria das campanhas de marketing, transformando o interesse em resultados tangíveis.

*   **Engajamento:** É uma métrica que mede a interação e o envolvimento do público com o conteúdo ou a marca, como curtidas, comentários, compartilhamentos e tempo de visualização. Indica o quão conectada a audiência está com a marca.

```python
def shuffle_column(df, column_name):
    """
    Embaralha uma coluna específica de um DataFrame sem alterar outras colunas.

    Args:
        df (pd.DataFrame): O DataFrame a ser modificado.
        column_name (str): O nome da coluna a ser embaralhada.

    Returns:
        pd.DataFrame: O DataFrame com a coluna especificada embaralhada.
    """
    df_shuffled = df.copy()
    df_shuffled[column_name] = df_shuffled[column_name].sample(frac=1, random_state=42).reset_index(drop=True)
    return df_shuffled

print("A função 'shuffle_column' foi definida.")
```

**Coluna 'Canais'**

A coluna 'Canais' no DataFrame `df_3` foi padronizada para garantir que cada linha contenha apenas um canal, mesmo quando originalmente havia múltiplos canais listados. Essa transformação foi realizada para simplificar a análise e garantir a consistência dos dados.

A lógica de seleção aplicada foi a seguinte:

*   **Para linhas com dois canais (ex: 'Facebook, Instagram')**: A seleção alternou sequencialmente entre o primeiro e o segundo canal com base na ordem de ocorrência de cada linha. Por exemplo, a primeira ocorrência de dois canais pegou o primeiro canal, a segunda ocorrência pegou o segundo, a terceira pegou o primeiro novamente, e assim por diante.

*   **Para linhas com três canais (ex: 'WhatsApp, Google, YouTube')**: A seleção ciclou entre o primeiro, o segundo e o terceiro canal, também com base na ordem de ocorrência de cada linha. Por exemplo, a primeira ocorrência de três canais pegou o primeiro, a segunda o segundo, a terceira o terceiro, a quarta o primeiro, e assim sucessivamente.

*   **Para linhas com apenas um canal**: O canal existente foi mantido sem alterações.

Esta padronização foi implementada utilizando a função `select_channel`, que gerencia os contadores internos (`count_two_channels` e `count_three_channels`) para aplicar a lógica de alternância e ciclagem de forma consistente em todo o DataFrame.

```python
count_two_channels = 0
count_three_channels = 0

print(f"count_two_channels initialized to: {count_two_channels}")
print(f"count_three_channels initialized to: {count_three_channels}")
```

```python
def select_channel(channel_string):
    global count_two_channels
    global count_three_channels

    channels = [c.strip() for c in channel_string.split(',')]
    num_channels = len(channels)

    if num_channels == 1:
        return channels[0]
    elif num_channels == 2:
        count_two_channels += 1
        return channels[(count_two_channels - 1) % 2]
    elif num_channels == 3:
        count_three_channels += 1
        return channels[(count_three_channels - 1) % 3]
    else:
        # For cases with more than 3 channels or empty strings, return the first one as default.
        return channels[0] if channels else ''

print("The 'select_channel' function has been defined.")
```

```python
# Aplicando a função
df_1['Canais'] = df_1['Canais'].apply(select_channel)
```
